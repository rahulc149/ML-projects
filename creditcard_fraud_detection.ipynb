```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Load dataset
creditcard_data = pd.read_csv("creditcard.csv")
```

```python
#First 5 rows of the dataset
creditcard_data.head()
```

```python
creditcard_data.tail()
```

```python
#Dataset info
creditcard_data.info()
```

```python
# Checking the missing values
creditcard_data.isnull().sum()
```

```python
# Distribution of legit and fraudulent transaction
creditcard_data['Class'].value_counts()
```

```python
# separating data for analysis
legit = creditcard_data[creditcard_data.Class==0]
fraud = creditcard_data[creditcard_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare values for both transcation
creditcard_data.groupby('Class').mean()
```

```python
# Taking sample dataset from exixting dataset
legit_sample = legit.sample(n=492)
```

```python
legit_sample
```

```python
# Concating two data sets 
new_ds = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_ds.head()
```

```python
new_ds.tail()
```

```python
new_ds['Class'].value_counts()
```

```python
new_ds.groupby('Class').mean()
```

```python
# Splitting data into features and target
x = new_ds.drop(columns='Class',axis=1)
y = new_ds['Class']
```

```python
x
```

```python
y
```

```python
# Split data into train and test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
# Model training
model = LogisticRegression()
```

```python
# training logistic regression model with training data
model.fit(x_train,y_train)
```

```python
# Accuracy score on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on training data: ',training_data_accuracy)
```

```python
# Accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on testing data: ',test_data_accuracy)
```
