```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Marks.csv")
```

```python
df.head()
```

```python
import numpy as np
array_data = df.values
array_data
```

```python
subjects = array_data[1:, 0]
cie1_scores =array_data[1:, 1].astype(int)
cie2_scores = array_data[1:, 2].astype(int)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(subjects, cie1_scores, marker='o', label='CIE-1', color='b')
plt.plot(subjects, cie2_scores, marker='*', label='CIE-2', color='g')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Internals vs. Subjects[RAHUL C]')
plt.legend()
```
